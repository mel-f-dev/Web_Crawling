```python
from konlpy.tag import Twitter
from collections import Counter
```

```python
file = open("유미의 세포들_comments.txt", mode = 'r', encoding = 'utf-8')
lists = file.readlines()
file.close()
lists
```

```python
import re

def clean_text(text):
    cleaned_text = re.sub('[a-zA-Z]' , '', text)
    cleaned_text_punc = re.sub('[\{\}\[\]\/?.,;:|\)*~`!^\-_+<>@\#$%&\\\=\(\'\"]',
                          '', cleaned_text)
    EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
    cleaned_text_emoji = EMOJI.sub(r'', cleaned_text_punc)
 
    return cleaned_text_emoji

re_comments =[]
for i in lists:
    re_text = clean_text(i)
    re_comments.append(re_text.rstrip())
```

```python
re_comments
```

```python
comments = [x for x in re_comments if x]
comments
```

```python
import csv

with open('yums_comments.csv', 'w', newline = '', encoding = 'utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(comments)
```

```python
import pandas as pd
df = pd.DataFrame(comments)
df.to_csv("", header = False, index = False)
```

```python
twitter = Twitter()
morphs = []

# 형태소 분석
for sentence in comments:
    morphs.append(twitter.pos(sentence))

print(morphs)
```

```python
noun_adj_adv_list = []

for sentence in morphs:
    for word, tag in sentence:
        if tag in ['Noun'] and ("것" not in word) and ("내" not in word)and ("나" not in word)and ("저" not in word) and ("개" not in word) and ("진짜" not in word) and ("거" not in word) and ("왜" not in word):
            noun_adj_adv_list .append(word)
print(noun_adj_adv_list)
```

```python
count = Counter(noun_adj_adv_list)
count
```

```python
words = dict(count.most_common())
```
